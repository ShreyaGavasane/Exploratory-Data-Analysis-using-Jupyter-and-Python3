### Ploting a bar graph of the salary figures of each organization.
As the dataset is very huge it is difficult to fit in the frame.

```python
df_temp = dframe1[['organization','annual_salary','ytd_gross_earnings','regular_earnings','overtime_earnings','other_earnings']]
df_temp.groupby('organization').nunique().plot(kind='bar', figsize=(15,4))
plt.show()
```

### Displaying the maximum and minimum annual salary present in dframe1. Also, displaying the count of maximum and minimum salary present. 
### Visualizing the annual salaries present in dframe1
We can also see the minimum and maximum salary and the the organization name in the ploted graph. 
As the dataset is really huge it is not able to fit in the visualisation frame.

```python
print("Maximum annual salary in Data set\n\n",dframe1[dframe1.annual_salary==dframe1.annual_salary.max()])
print("\n\nCount of number of rows having Maximum annual salary : ",len(dframe1[dframe1.annual_salary==dframe1.annual_salary.max()]))
print("\n------------------------------------------------------------------------------------\n")
print("Minimum annual salary in Data set\n\n",dframe1[dframe1.annual_salary==dframe1.annual_salary.min()])
print("\n\nCount of number of rows having Minimum annual salary : ",len(dframe1[dframe1.annual_salary==dframe1.annual_salary.min()]))


print("\n-------------------------------------------------------------------------------------\n")
dframe1.plot(x='organization',y='annual_salary',style='o', figsize=(40,10))
plt.tick_params(labelsize=14)
plt.xlabel('organization',fontsize=14)
plt.ylabel('Annual Salaries',fontsize=14)
plt.show()
```

### Visualizing the count of each organization record using pie chart in percentage.
As the dataset is very huge the name and numbers are overlapping each other

```python
labels = dframe1['organization'].value_counts().index
values = dframe1['organization'].value_counts().values
explode = (0,0,0.8,0.8)
fig1, ax1 = plt.subplots(figsize=(8, 8))
ax1.pie(values, labels=labels, autopct='%1.1f%%',
        #shadow=True,
        startangle=90)
ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

plt.show()
```

### Taking the aggregate of annual_salary, regular earnings, overtime earnings, other earnings, and gross earnings of all the organizations using groupby and displaying it.

```python
aggr = dframe1.groupby(['organization']).agg(np.mean)
print(aggr)
```

### Displaying maximum and minimum aggregate annual salaries. Also, displaying the count of number of rows having maximum and minimum of the same.

```python
print("Max aggregate of annual salary : \n\n",aggr[aggr.annual_salary==aggr.annual_salary.max()] )
print("\nCount : ",len(aggr[aggr.annual_salary==aggr.annual_salary.max()]))
print("---------------------------------------------------------------------\n\n")
print("Min aggregate of annual salary : \n\n",aggr[aggr.annual_salary==aggr.annual_salary.min()])
print("\nCount : ",len(aggr[aggr.annual_salary==aggr.annual_salary.min()]))
```
